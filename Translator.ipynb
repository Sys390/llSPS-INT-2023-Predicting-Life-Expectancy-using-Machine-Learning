### Text to translate
We take a text from an article titled (in English): "German Firms Investigated Over Possible Tax Evasions..."

```python
text_input = "Dem Handelsblatt zufolge Finanzminister von Nordrhein-Westfalen Norbert Walter-Borjans sagte, \
dass eine anonyme Quelle eine Speicherkarte mit Informationen \u00fcber bis zu 70.000 Unternehmen in Malta mit \
fast 2.000 von ihnen mit deutschen Unternehmen verkn\u00fcpft \u00fcbergab. Insgesamt 85 % von ihnen wurden \
Berichten zufolge nicht mit deutschen Steuerbeh\u00f6rden registriert."
```

### Identify the language of the text
The language translator can identify over 60 languages

```python
language = translator.identify(text_input).get_result()
# print(json.dumps(language, indent=2)) 
language["languages"][0:3] # show the top three possible languages
```

### Translate the text to English
We've identified that the most likely source language is `de` (German)so we can use the API to translate to `en` (English)

For the news environment, this service can translate back and forth between English and seven other languages

```python
english_text = translator.translate(text_input, source="de", target="en").get_result()

# The entire output includes character_count, word_count, and translations
# print(json.dumps(english_text, indent=2))
english_text["translations"][0]["translation"]
```
